Trends in revenue and expenses by category and season

Seasonal Trends: Average Revenue and Expenses

Monthly Revenue, Expenses, and Profit

```python
df['Revenue'] = df['Total']
df['Expenses'] = df['Revenue'] * np.random.uniform(0.6, 0.8, size=len(df))
df['Profit'] = df['Revenue'] - df['Expenses']

monthly_data = df.groupby(pd.Grouper(key='Date', freq='M')).agg({'Revenue': 'sum', 'Expenses': 'sum', 'Profit': 'sum'}).reset_index()

# Plot trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Date'], monthly_data['Revenue'], label='Revenue')
plt.plot(monthly_data['Date'], monthly_data['Expenses'], label='Expenses')
plt.plot(monthly_data['Date'], monthly_data['Profit'], label='Profit')
plt.legend()
plt.title('Monthly Revenue, Expenses, and Profit')
plt.show()
```

Model Comparision: ARIMA, Prophet, XGBoost

```python
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ARIMA
arima_series = monthly.set_index(monthly['Month'].dt.to_timestamp())['gross income'] # Using 'gross income' instead of 'Profit' as 'Profit' was created using random values and not present in 'monthly'
train_arima = arima_series[:-3]
test_arima = arima_series[-3:]
model_arima = SARIMAX(train_arima, order=(1,1,1), seasonal_order=(0,0,0,0))
result_arima = model_arima.fit()

# get integer positions of start and end
start_idx = len(train_arima)
end_idx = len(train_arima) + len(test_arima) - 1

forecast_arima = result_arima.predict(start=start_idx, end=end_idx, typ='levels')
```

```python
# Prophet
prophet_df = monthly[['Month', 'gross income']].rename(columns={"Month":"ds", "gross income":"y"})
prophet_df['ds'] = prophet_df['ds'].dt.to_timestamp() # Convert Period to Timestamp
train_prophet = prophet_df
test_prophet = prophet_df

m_prophet = Prophet()
m_prophet.fit(train_prophet)
future = m_prophet.make_future_dataframe(periods=3, freq="M")
forecast_prophet = m_prophet.predict(future)
forecast_prophet_values = forecast_prophet[['ds','yhat']].tail(3)
```

```python
# XGBoost
xgb_data = monthly.copy()
xgb_data['Month'] = xgb_data['Month'].dt.to_timestamp().dt.month # Use 'Month' column and convert to datetime
xgb_data['Lag1'] = xgb_data['gross income'].shift(1) # Use 'gross income'
xgb_data['Lag2'] = xgb_data['gross income'].shift(2) # Use 'gross income'
xgb_data.dropna(inplace=True)

train_xgb = xgb_data[:-3]
test_xgb = xgb_data[-3:]
X_train = train_xgb[['Month', 'Lag1', 'Lag2']]
y_train = train_xgb['gross income'] # Use 'gross income' for target
X_test = test_xgb[['Month', 'Lag1', 'Lag2']]
y_test = test_xgb['gross income'] # Use 'gross income' for target

model_xgb = XGBRegressor()
model_xgb.fit(X_train, y_train)
pred_xgb = model_xgb.predict(X_test)
```

```python
forecast_arima_values = forecast_arima
forecast_prophet_values = forecast_prophet_values['yhat']
forecast_xgb_values = pred_xgb
actual_values = test_arima
```

```python
# evaluation metrics
mae_arima = mean_absolute_error(actual_values, forecast_arima_values)
rmse_arima = np.sqrt(mean_squared_error(actual_values, forecast_arima_values))

mae_prophet = mean_absolute_error(actual_values, forecast_prophet_values)
rmse_prophet = np.sqrt(mean_squared_error(actual_values, forecast_prophet_values))

mae_xgb = mean_absolute_error(y_test, forecast_xgb_values)
rmse_xgb = np.sqrt(mean_squared_error(y_test, forecast_xgb_values))

print(" Model Evaluation (MAE & RMSE)")
print(f"ARIMA   - MAE: {mae_arima:.2f}, RMSE: {rmse_arima:.2f}")
print(f"Prophet - MAE: {mae_prophet:.2f}, RMSE: {rmse_prophet:.2f}")
print(f"XGBoost - MAE: {mae_xgb:.2f}, RMSE: {rmse_xgb:.2f}")
```
